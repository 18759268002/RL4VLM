```python
import matplotlib.pyplot as plt
import gymnasium as gym
import gym_cards
import numpy as np
import random
```

```python
env = gym.make('gym_cards/WebClickEnv-v1')
obs, info = env.reset()
plt.imshow(obs)
```

```python
action = np.array([random.uniform(0, 300), random.uniform(0, 300)], dtype=np.float32)
print(action)
obs, reward, terminated,truncated, info = env.step(action)
plt.imshow(obs)
print(reward, terminated, truncated, info)
```
